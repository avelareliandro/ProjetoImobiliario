```python
import pandas as pd
url = 'https://raw.githubusercontent.com/avelareliandro/ProjetoImobiliario/refs/heads/main/alunos.csv'
df_alunos = pd.read_csv(url)
df_alunos.head()
```

```python
df_alunos.shape
```

```python
df_alunos.info()
# A função info() é usada para obter uma visão geral do DataFrame, incluindo o número de entradas, o número de colunas, os tipos de dados de cada coluna e a quantidade de valores não nulos em cada coluna. Isso é útil para entender a estrutura dos dados e identificar possíveis problemas, como valores ausentes ou tipos de dados incorretos.
```

```python
df_alunos.shape
# A função shape é usada para obter as dimensões do DataFrame, ou seja, o número de linhas e colunas. O resultado é uma tupla onde o primeiro elemento representa o número de linhas e o segundo elemento representa o número de colunas. Isso é útil para entender a quantidade de dados que estamos lidando e para verificar se o DataFrame foi carregado corretamente.
```

```python
df_alunos[['Nome', 'Aprovado']]
# A sintaxe df_alunos[['Nome', 'Aprovado']] é usada para selecionar apenas as colunas "Nome" e "Aprovado" do DataFrame df_alunos. O resultado será um novo DataFrame contendo apenas essas duas colunas, o que pode ser útil para focar em informações específicas ou para realizar análises mais direcionadas.
```

```python
df_alunos.head(7)
# A função head() é usada para exibir as primeiras linhas de um DataFrame. Ao passar o argumento 7, estamos solicitando que sejam exibidas as primeiras 7 linhas do DataFrame df_alunos. Isso é útil para obter uma visão rápida dos dados e verificar se o DataFrame foi carregado corretamente, além de permitir uma inspeção inicial dos valores contidos nas colunas.
```

```python
df_alunos.tail()
# A função tail() é usada para exibir as últimas linhas de um DataFrame. Por padrão, ela exibe as últimas 5 linhas, mas você pode passar um argumento para especificar quantas linhas deseja ver. Isso é útil para obter uma visão rápida do final dos dados e verificar se o DataFrame foi carregado corretamente, além de permitir uma inspeção inicial dos valores contidos nas colunas.
```
